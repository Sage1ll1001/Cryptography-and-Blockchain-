```python
import hashlib
import rsa

# ----------------------------
# VEHICLE REGISTRATION SYSTEM
# ----------------------------
vehicles = {}  # {number_plate: {"owner": ..., "model": ...}}

def register_vehicle():
    number_plate = input("Enter vehicle number plate: ").strip().upper()
    if number_plate in vehicles:
        print("‚ùå Vehicle already registered!")
        return
    owner = input("Enter owner name: ").strip()
    model = input("Enter vehicle model: ").strip()
    if not owner or not model:
        print("‚ùå Invalid input. Try again.")
        return
    vehicles[number_plate] = {"owner": owner, "model": model}
    print(f"‚úÖ Vehicle {number_plate} registered successfully!")


def retrieve_vehicle():
    number_plate = input("Enter vehicle number plate to retrieve: ").strip().upper()
    if number_plate in vehicles:
        details = vehicles[number_plate]
        print("\nüöó Vehicle Details:")
        print(f"Number Plate: {number_plate}")
        print(f"Owner: {details['owner']}")
        print(f"Model: {details['model']}\n")
    else:
        print("‚ùå Vehicle not found.")


# ----------------------------
# SHA-256 HASHING
# ----------------------------
def sha256_hash():
    message = input("Enter message to hash: ")
    hash_value = hashlib.sha256(message.encode()).hexdigest()
    print(f"\nüîí SHA-256 Hash:\n{hash_value}\n")


# ----------------------------
# DIGITAL SIGNATURE
# ----------------------------
def digital_signature():
    print("üîë Generating RSA public‚Äìprivate key pair...")
    (public_key, private_key) = rsa.newkeys(512)
    print("‚úÖ Keys generated successfully.\n")

    message = input("Enter message to sign: ").encode()

    # Sign message using private key
    signature = rsa.sign(message, private_key, 'SHA-256')
    print("\nüìù Message signed successfully!")

    # Verify signature using public key
    try:
        rsa.verify(message, signature, public_key)
        print("‚úÖ Signature verified successfully ‚Äî message is authentic!\n")
    except rsa.VerificationError:
        print("‚ùå Signature verification failed!\n")


# ----------------------------
# MAIN MENU
# ----------------------------
def menu():
    while True:
        print("========== CRYPTOGRAPHY & BLOCKCHAIN APP ==========")
        print("1Ô∏è‚É£  Generate SHA-256 Hash")
        print("2Ô∏è‚É£  Create & Verify Digital Signature")
        print("3Ô∏è‚É£  Register Vehicle")
        print("4Ô∏è‚É£  Retrieve Vehicle Details")
        print("5Ô∏è‚É£  Exit")
        print("===================================================")

        choice = input("Enter your choice (1-5): ").strip()

        if choice == "1":
            sha256_hash()
        elif choice == "2":
            digital_signature()
        elif choice == "3":
            register_vehicle()
        elif choice == "4":
            retrieve_vehicle()
        elif choice == "5":
            print("üëã Exiting program. Goodbye!")
            break
        else:
            print("‚ùå Invalid choice! Please enter a number between 1‚Äì5.\n")


# ----------------------------
# RUN THE PROGRAM
# ----------------------------
if __name__ == "__main__":
    menu()
```
